```python
def clock(a, b, c):
    u = b - a
    v = c - b
    return u.real * v.imag - u.imag * v.real < 0
 
 
n = int(input())
if n < 4 or n > 1000:
    print("N is out of index")
    exit()
p = []
for i in range(n + 1):
    x, y = map(int, input().split())
    if -10000 > x or x > 10000 or -10000 > y or y > 10000:
        print("x or y is out of index")
        exit()
    p.append(complex(x, y))
 
cnt = 0
for i in range(1, n):
    if not clock(p[i - 1], p[i], p[i + 1]):
        cnt += 1
 
print(cnt)
```

–¢–æ–ª—å–∫–æ —Å–æ—Ç—Ä—É–¥–Ω–∏–∫ 1 –æ—Å—Ç–∞–ª—Å—è —Å –ø—Ä–∞–≤–æ–º –≥–æ–ª–æ—Å–∞, –ø–æ—ç—Ç–æ–º—É –≥–æ–ª–æ—Å–æ–≤–∞–Ω–∏–µ –∑–∞–∫–∞–Ω—á–∏–≤–∞–µ—Ç—Å—è –ø—Ä–∏–Ω—è—Ç–∏–µ–º —Ä–µ—à–µ–Ω–∏—è –≤ –ø–æ–ª—å–∑—É —Ñ—Ä–∞–∫—Ü–∏–∏ –¥–µ–ø—É–±–ª–∏–∫–∞–Ω—Ü–µ–≤.

```python
n = int(input())
if 1 <= n <= 200000:
    print("n is out of range")
    exit()
req = input()
while req.count("R") > 0 and req.count("D") > 0:
    for i in range(len(req)):
        if req[i] == "D":
            req = req.replace("R", "N", 1)
        if req[i] == "R":
            req = req.replace("D", "N", 1)
if req.count("R") == 0:
    print("D")
else:
    print("R")
```

## –ó–∞–¥–∞—á–∞ 3
–ê–ª—ë–Ω–∞ —Ä–∞–±–æ—Ç–∞–µ—Ç –≤ –∞—ç—Ä–æ–ø–æ—Ä—Ç—É –ú–µ–≥–∞–ø–æ–ª–∏—Å–∞, —Å–æ—Å—Ç–∞–≤–ª—è—è —Ä–∞—Å–ø–∏—Å–∞–Ω–∏–µ –æ—Ç–ø—Ä–∞–≤–ª–µ–Ω–∏—è —Ä–µ–π—Å–æ–≤. –°–µ–≥–æ–¥–Ω—è –¥–æ–ª–∂–Ω—ã –æ—Ç–ø—Ä–∞–≤–∏—Ç—å—Å—è n —Ä–µ–π—Å–æ–≤, i-–π –∏–∑ –∫–æ—Ç–æ—Ä—ã—Ö –¥–æ–ª–∂–µ–Ω –≤—ã–ª–µ—Ç–µ—Ç—å –≤ –º–∏–Ω—É—Ç—É i.

–ö–∞–∫ –≤—ã –ø–æ–º–Ω–∏—Ç–µ –∏–∑ –∑–∞–¥–∞—á–∏ –≤—á–µ—Ä–∞—à–Ω–µ–≥–æ –¥–Ω—è, –∞—ç—Ä–æ–ø–æ—Ä—Ç –ú–µ–≥–∞–ø–æ–ª–∏—Å–∞ —è–≤–ª—è–µ—Ç—Å—è –æ—Å–Ω–æ–≤–Ω—ã–º —Ç—Ä–∞–Ω—Å–ø–æ—Ä—Ç–Ω—ã–º —É–∑–ª–æ–º –≤ –ú–µ—Ç—Ä–æ–ø–æ–ª–∏–∏, –∞ –≤ –≤—ã—Å–æ–∫–æ–Ω–∞–≥—Ä—É–∂–µ–Ω–Ω—ã—Ö –∞—ç—Ä–æ–ø–æ—Ä—Ç–∞—Ö –Ω–µ—Ä–µ–¥–∫–æ —Å–ª—É—á–∞—é—Ç—Å—è –Ω–∞–∫–ª–∞–¥–∫–∏. –†–æ–≤–Ω–æ —Ç–∞–∫ –∏ –ø—Ä–æ–∏–∑–æ—à–ª–æ —Å–µ–≥–æ–¥–Ω—è ‚Äî –∏–∑-–∑–∞ —Ç–µ—Ö–Ω–∏—á–µ—Å–∫–∏—Ö –Ω–µ–ø–æ–ª–∞–¥–æ–∫ –≤ —Ç–µ—á–µ–Ω–∏–µ –ø–µ—Ä–≤—ã—Ö k –º–∏–Ω—É—Ç –Ω–∏ –æ–¥–∏–Ω —Ä–µ–π—Å –Ω–µ —Å–º–æ–≥ –≤—ã–ª–µ—Ç–µ—Ç—å –∏–∑ –∞—ç—Ä–æ–ø–æ—Ä—Ç–∞.

–¢–µ–ø–µ—Ä—å –≤—Å–µ –∑–∞–ø–ª–∞–Ω–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ n —Ä–µ–π—Å–æ–≤ –¥–æ–ª–∂–Ω—ã –≤—ã–ª–µ—Ç–µ—Ç—å –≤ —Ä–∞–∑–ª–∏—á–Ω—ã–µ –º–∏–Ω—É—Ç—ã –æ—Ç (k‚Äâ+‚Äâ1)-–π –¥–æ (k‚Äâ+‚Äân)-–π –≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ. –û–¥–Ω–∞–∫–æ —Ä–µ–π—Å—ã –Ω–µ –æ–±—è–∑–∞–Ω—ã –≤—ã–ª–µ—Ç–∞—Ç—å –≤ –∏—Å—Ö–æ–¥–Ω–æ–º –ø–æ—Ä—è–¥–∫–µ ‚Äî –ê–ª—ë–Ω–∞ –º–æ–∂–µ—Ç —Å–æ—Å—Ç–∞–≤–∏—Ç—å –ª—é–±–æ–µ —Ä–∞—Å–ø–∏—Å–∞–Ω–∏–µ –æ—Ç–ø—Ä–∞–≤–ª–µ–Ω–∏—è —Ä–µ–π—Å–æ–≤. –ü—Ä–∏ —ç—Ç–æ–º –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å –≤—ã–ø–æ–ª–Ω–µ–Ω–æ –≤–∞–∂–Ω–æ–µ —É—Å–ª–æ–≤–∏–µ: –Ω–∏ –æ–¥–∏–Ω —Ä–µ–π—Å –Ω–µ –º–æ–∂–µ—Ç –≤—ã–ª–µ—Ç–µ—Ç—å —Ä–∞–Ω—å—à–µ —Å–≤–æ–µ–≥–æ –∏–∑–Ω–∞—á–∞–ª—å–Ω–æ –∑–∞–ø–ª–∞–Ω–∏—Ä–æ–≤–∞–Ω–Ω–æ–≥–æ –≤—Ä–µ–º–µ–Ω–∏ –≤—ã–ª–µ—Ç–∞.

–ê–ª—ë–Ω–∞ –∑–Ω–∞–µ—Ç, —á—Ç–æ –∑–∞–¥–µ—Ä–∂–∫–∞ –≤—ã–ª–µ—Ç–∞ i-–≥–æ —Ä–µ–π—Å–∞ –Ω–∞ –æ–¥–Ω—É –º–∏–Ω—É—Ç—É —Å—Ç–æ–∏—Ç –∞—ç—Ä–æ–ø–æ—Ä—Ç—É ci –±—É—Ä–ª–µ–π. –ü–æ–º–æ–≥–∏—Ç–µ –ê–ª—ë–Ω–µ –æ–ø—Ä–µ–¥–µ–ª–∏—Ç—å, –≤ –∫–∞–∫–æ–º –ø–æ—Ä—è–¥–∫–µ —Å–ª–µ–¥—É–µ—Ç –≤—ã–ª–µ—Ç–∞—Ç—å —Ä–µ–π—Å–∞–º, —á—Ç–æ–±—ã —Å—É–º–º–∞—Ä–Ω–∞—è —Å—Ç–æ–∏–º–æ—Å—Ç—å –∑–∞–¥–µ—Ä–∂–∫–∏ –æ–∫–∞–∑–∞–ª–∞—Å—å –º–∏–Ω–∏–º–∞–ª—å–Ω–æ –≤–æ–∑–º–æ–∂–Ω–æ–π.

–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
–í –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö –Ω–∞—Ö–æ–¥—è—Ç—Å—è –¥–≤–∞ —Ü–µ–ª—ã—Ö —á–∏—Å–ª–∞ n –∏ k (1‚Äâ‚â§‚Äâk‚Äâ‚â§‚Äân‚Äâ‚â§‚Äâ300‚Äâ000), –≥–¥–µ n ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ä–µ–π—Å–æ–≤, –∞ k ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –º–∏–Ω—É—Ç, –≤ —Ç–µ—á–µ–Ω–∏–µ –∫–æ—Ç–æ—Ä—ã—Ö —Ä–µ–π—Å—ã –Ω–µ –≤—ã–ª–µ—Ç–∞–ª–∏ –∏–∑ –∞—ç—Ä–æ–ø–æ—Ä—Ç–∞.

–í–æ –≤—Ç–æ—Ä–æ–π —Å—Ç—Ä–æ–∫–µ –Ω–∞—Ö–æ–¥—è—Ç—Å—è n —á–∏—Å–µ–ª c1,‚Äâc2,‚Äâ...,‚Äâcn (1‚Äâ‚â§‚Äâci‚Äâ‚â§‚Äâ107), –≥–¥–µ ci ‚Äî —Å—Ç–æ–∏–º–æ—Å—Ç—å –∑–∞–¥–µ—Ä–∂–∫–∏ i-–≥–æ —Ä–µ–π—Å–∞ –Ω–∞ –æ–¥–Ω—É –º–∏–Ω—É—Ç—É.

–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
–í –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ –≤—ã–≤–µ–¥–∏—Ç–µ –º–∏–Ω–∏–º–∞–ª—å–Ω—É—é –≤–æ–∑–º–æ–∂–Ω—É—é —Å—É–º–º–∞—Ä–Ω—É—é —Å—Ç–æ–∏–º–æ—Å—Ç—å –∑–∞–¥–µ—Ä–∂–∫–∏ –≤—Å–µ—Ö —Ä–µ–π—Å–æ–≤.

–í–æ –≤—Ç–æ—Ä–æ–π —Å—Ç—Ä–æ–∫–µ –≤—ã–≤–µ–¥–∏—Ç–µ n —Ä–∞–∑–ª–∏—á–Ω—ã—Ö —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª t1,‚Äât2,‚Äâ...,‚Äâtn (k‚Äâ+‚Äâ1‚Äâ‚â§‚Äâti‚Äâ‚â§‚Äâk‚Äâ+‚Äân), –≥–¥–µ ti –æ–∑–Ω–∞—á–∞–µ—Ç –≤—Ä–µ–º—è –≤—ã–ª–µ—Ç–∞ i-–≥–æ —Ä–µ–π—Å–∞. –ï—Å–ª–∏ —Ä–∞—Å–ø–∏—Å–∞–Ω–∏–π —Å –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–π —Å—Ç–æ–∏–º–æ—Å—Ç—å—é –Ω–µ—Å–∫–æ–ª—å–∫–æ, —Ä–∞–∑—Ä–µ—à–∞–µ—Ç—Å—è –≤—ã–≤–µ—Å—Ç–∏ –ª—é–±–æ–µ –∏–∑ –Ω–∏—Ö.

–ü—Ä–∏–º–µ—Ä\
–≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\
5 2\
4 2 1 10 2\
–≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ\
–†–∞—Å—Å–º–æ—Ç—Ä–∏–º –ø—Ä–∏–º–µ—Ä –∏–∑ —É—Å–ª–æ–≤–∏—è. –ï—Å–ª–∏ —Å–¥–≤–∏–Ω—É—Ç—å —Ä–µ–π—Å—ã –Ω–∞ 2 –º–∏–Ω—É—Ç—ã, —Å–æ—Ö—Ä–∞–Ω–∏–≤ –∏—Ö –ø–æ—Ä—è–¥–æ–∫ –≤—ã–ª–µ—Ç–∞, —Ç–æ —Å—É–º–º–∞—Ä–Ω–∞—è —Å—Ç–æ–∏–º–æ—Å—Ç—å –∑–∞–¥–µ—Ä–∂–∫–∏ –≤—Å–µ—Ö —Ä–µ–π—Å–æ–≤ —Å–æ—Å—Ç–∞–≤–∏—Ç (3‚Äâ-‚Äâ1)¬∑4‚Äâ+‚Äâ(4‚Äâ-‚Äâ2)¬∑2‚Äâ+‚Äâ(5‚Äâ-‚Äâ3)¬∑1‚Äâ+‚Äâ(6‚Äâ-‚Äâ4)¬∑10‚Äâ+‚Äâ(7‚Äâ-‚Äâ5)¬∑2‚Äâ=‚Äâ38 –±—É—Ä–ª–µ–π.

–ï—Å–ª–∏ –∂–µ –∏–∑–º–µ–Ω–∏—Ç—å —Ä–∞—Å–ø–∏—Å–∞–Ω–∏–µ —Ç–∞–∫, –∫–∞–∫ –ø–æ–∫–∞–∑–∞–Ω–æ –≤ –æ—Ç–≤–µ—Ç–µ, —Ç–æ —Å—É–º–º–∞—Ä–Ω–∞—è —Å—Ç–æ–∏–º–æ—Å—Ç—å —Å–æ—Å—Ç–∞–≤–∏—Ç (3‚Äâ-‚Äâ1)¬∑4‚Äâ+‚Äâ(6‚Äâ-‚Äâ2)¬∑2‚Äâ+‚Äâ(7‚Äâ-‚Äâ3)¬∑1‚Äâ+‚Äâ(4‚Äâ-‚Äâ4)¬∑10‚Äâ+‚Äâ(5‚Äâ-‚Äâ5)¬∑2‚Äâ=‚Äâ20 –±—É—Ä–ª–µ–π.

```python
import itertools
k, n = map(int, input().split())
if k > 300000 or k < 1 or n > 300000 or n < 1:
    print("k or n is out of range")
    exit()
data = list(map(int, input().split()))
for item in data:
    if item < 1 or item > 107:
        print("C is out of range")
        exit()
min_v = []
min_value = 99999999999999
v = []
for i in range(k):
    v.append(i + n + 1)
variants = list(itertools.permutations(v, k))
for variant in variants:
    summ = 0
    for i in range(len(variant)):
        summ += (variant[i] - i - 1) * data[i]
        if variant[i] - i - 1 < 0:
            summ = 9999999999999999
            break
    if summ <= min_value:
        min_value = summ
        min_v = variant
print(min_value)
for item in min_v:
    print(item, end=" ")
```

–°—É–º–º–∞—Ä–Ω–∞—è —Å—Ç–æ–∏–º–æ—Å—Ç—å 3+6+12+24+3+48+75=171. –ù–µ–ª—å–∑—è –ø–æ–ª—É—á–∏—Ç—å –º–µ–Ω—å—à—É—é —Å—Ç–æ–∏–º–æ—Å—Ç—å.

```python
def solve():
    n, a, b = map(int, input().split())
    if n < 1 or n > 210:
        print("n is out of range")
        exit()
    if a < 1 or a > 105:
        print("a is out of range")
        exit()
    if b < 1 or b > 105:
        print("b is out of range")
        exit()
    xs = list(map(int, input().split()))
    for item in xs:
        if item < 1 or item > 108:
            print("x is out of range")
            exit()
    c = 0
    ans = sum(xs) * b
    for i in range(n):
        x = xs[-i - 1]
        ans = min(ans, x * (a + b) + (c - x * i) * b)
        c += x

    print(ans)

t = int(input())
if t < 1 or t > 1000:
    print("t is out of range")
    exit()
for _ in range(t):
    solve()
```

–ü–æ–º–æ–≥–∏—Ç–µ –ú–∞–ª–µ–∫—É –æ—Ç–∫—Ä—ã—Ç—å –¥–≤–µ—Ä—å, —É–∫–∞–∑–∞–≤ –µ–º—É, —Å–∫–æ–ª—å–∫–∏–º–∏ —Å–∏–º–≤–æ–ª–∞–º–∏ ')' –Ω–∞–¥–æ –∑–∞–º–µ–Ω–∏—Ç—å –∫–∞–∂–¥—ã–π —Å–∏–º–≤–æ–ª '#'.

–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
–ï—Å–ª–∏ –Ω–µ–≤–æ–∑–º–æ–∂–Ω–æ –∑–∞–º–µ–Ω–∏—Ç—å —Å–∏–º–≤–æ–ª—ã '#' —Ç–∞–∫, —á—Ç–æ–±—ã –ø–æ–ª—É—á–∏–ª–∞—Å—å –∫—Ä–∞—Å–∏–≤–∞—è —Å—Ç—Ä–æ–∫–∞, –≤—ã–≤–µ–¥–∏—Ç–µ ‚Äâ-‚Äâ1. –í –ø—Ä–æ—Ç–∏–≤–Ω–æ–º —Å–ª—É—á–∞–µ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ —Å–∏–º–≤–æ–ª–∞ '#' –≤—ã–≤–µ–¥–∏—Ç–µ –≤ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–µ —á–∏—Å–ª–æ ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–∏–º–≤–æ–ª–æ–≤ ')', –∫–æ—Ç–æ—Ä—ã–º–∏ –Ω–∞–¥–æ –∑–∞–º–µ–Ω–∏—Ç—å —ç—Ç–æ—Ç —Å–∏–º–≤–æ–ª.

–ï—Å–ª–∏ –≤–æ–∑–º–æ–∂–Ω—ã—Ö –æ—Ç–≤–µ—Ç–æ–≤ –Ω–µ—Å–∫–æ–ª—å–∫–æ, —Ä–∞–∑—Ä–µ—à–∞–µ—Ç—Å—è –≤—ã–≤–µ—Å—Ç–∏ –ª—é–±–æ–π.

–ü—Ä–∏–º–µ—Ä—ã\
–≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\

–≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\
1\
2

```python
s = [*input()]
if len(s) < 1 or len(s) > 105:
    print("s is out of range")
    exit()
ans = []
t = h = 0
 
for i in range(len(s)):
    t += 1 if s[i] == '(' else -1
    if t < 0:
        print(-1)
        exit()
    if s[i] == '#':
        s[i] = ')'
        ans.append(1)
        h = i
 
ans[-1] += t
s = s[:h] + [')'] * t + s[h:]
 
t = 0
 
for i in s:
    t += 1 if i == '(' else -1
    if t < 0:
        print(-1)
        exit()
 
print(*ans, sep='\n')
```

## –ó–∞–¥–∞—á–∞ 7
–°–∫–æ–±–æ—á–Ω–æ–π –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å—é –Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è —Å—Ç—Ä–æ–∫–∞, —Å–æ—Å—Ç–æ—è—â–∞—è —Ç–æ–ª—å–∫–æ –∏–∑ —Å–∏–º–≤–æ–ª–æ–≤ ¬´(¬ª –∏ ¬´)¬ª.

–ü—Ä–∞–≤–∏–ª—å–Ω–æ–π —Å–∫–æ–±–æ—á–Ω–æ–π –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å—é –Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è —Å–∫–æ–±–æ—á–Ω–∞—è –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å, –∫–æ—Ç–æ—Ä—É—é –º–æ–∂–Ω–æ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞—Ç—å –≤ –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ–µ –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–µ –≤—ã—Ä–∞–∂–µ–Ω–∏–µ –ø—É—Ç–µ–º –≤—Å—Ç–∞–≤–æ–∫ –º–µ–∂–¥—É –µ–µ —Å–∏–º–≤–æ–ª–∞–º–∏ —Å–∏–º–≤–æ–ª–æ–≤ ¬´1¬ª –∏ ¬´+¬ª. –ù–∞–ø—Ä–∏–º–µ—Ä, —Å–∫–æ–±–æ—á–Ω—ã–µ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ ¬´()()¬ª, ¬´(())¬ª ‚Äî –ø—Ä–∞–≤–∏–ª—å–Ω—ã–µ (–ø–æ–ª—É—á–µ–Ω–Ω—ã–µ –≤—ã—Ä–∞–∂–µ–Ω–∏—è: ¬´(1)+(1)¬ª, ¬´((1+1)+1)¬ª), –∞ ¬´)(¬ª –∏ ¬´(¬ª ‚Äî –Ω–µ—Ç.

–í–∞–º –∑–∞–¥–∞–Ω–æ ùëõ —Å–∫–æ–±–æ—á–Ω—ã—Ö –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–µ–π ùë†1,ùë†2,‚Ä¶,ùë†ùëõ. –ù–∞–π–¥–∏—Ç–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–∞—Ä ùëñ,ùëó(1‚â§ùëñ,ùëó‚â§ùëõ), —Ç–∞–∫–∏—Ö —á—Ç–æ —Å–∫–æ–±–æ—á–Ω–∞—è –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å ùë†ùëñ+ùë†ùëó —è–≤–ª—è–µ—Ç—Å—è –ø—Ä–∞–≤–∏–ª—å–Ω–æ–π —Å–∫–æ–±–æ—á–Ω–æ–π –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å—é. –û–ø–µ—Ä–∞—Ü–∏—è + –æ–∑–Ω–∞—á–∞–µ—Ç –∫–æ–Ω–∫–∞—Ç–µ–Ω–∞—Ü–∏—é, —Ç.‚Äâ–µ. "()(" + ")()" = "()()()".

–ï—Å–ª–∏ ùë†ùëñ+ùë†ùëó –∏ ùë†ùëó+ùë†ùëñ —è–≤–ª—è—é—Ç—Å—è –ø—Ä–∞–≤–∏–ª—å–Ω—ã–º–∏ —Å–∫–æ–±–æ—á–Ω—ã–º–∏ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—è–º–∏ –∏ ùëñ‚â†ùëó, –≤ –æ—Ç–≤–µ—Ç–µ –¥–æ–ª–∂–Ω—ã —É—á–∏—Ç—ã–≤–∞—Ç—å—Å—è –æ–±–µ –ø–∞—Ä—ã (ùëñ,ùëó) –∏ (ùëó,ùëñ). –¢–∞–∫–∂–µ –µ—Å–ª–∏ —Å—Ç—Ä–æ–∫–∞ ùë†ùëñ+ùë†ùëñ —è–≤–ª—è–µ—Ç—Å—è –ø—Ä–∞–≤–∏–ª—å–Ω–æ–π —Å–∫–æ–±–æ—á–Ω–æ–π –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å—é, –ø–∞—Ä–∞ (ùëñ,ùëñ) –¥–æ–ª–∂–Ω–∞ —É—á–∏—Ç—ã–≤–∞—Ç—å—Å—è –≤ –æ—Ç–≤–µ—Ç–µ.

–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
–í –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ –∑–∞–¥–∞–Ω–æ —á–∏—Å–ª–æ ùëõ(1‚â§ùëõ‚â§3‚ãÖ105) ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–∫–æ–±–æ—á–Ω—ã—Ö –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–µ–π. –í —Å–ª–µ–¥—É—é—â–∏—Ö ùëõ —Å—Ç—Ä–æ–∫–∞—Ö —Å–æ–¥–µ—Ä–∂–∞—Ç—Å—è —Å–∫–æ–±–æ—á–Ω—ã–µ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ ‚Äî –Ω–µ–ø—É—Å—Ç—ã–µ —Å—Ç—Ä–æ–∫–∏, —Å–æ—Å—Ç–æ—è—â–∏–µ —Ç–æ–ª—å–∫–æ –∏–∑ —Å–∏–º–≤–æ–ª–æ–≤ ¬´(¬ª –∏ ¬´)¬ª. –°—É–º–º–∞ –¥–ª–∏–Ω –≤—Å–µ—Ö —Å–∫–æ–±–æ—á–Ω—ã—Ö –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–µ–π –Ω–µ –ø—Ä–µ–≤–æ—Å—Ö–æ–¥–∏—Ç 3‚ãÖ105.

–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
–í –µ–¥–∏–Ω—Å—Ç–≤–µ–Ω–Ω–æ–π —Å—Ç—Ä–æ–∫–µ –≤—ã–≤–µ–¥–∏—Ç–µ —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–∞—Ä ùëñ,ùëó(1‚â§ùëñ,ùëó‚â§ùëõ), —Ç–∞–∫–∏—Ö, —á—Ç–æ —Å–∫–æ–±–æ—á–Ω–∞—è –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å ùë†ùëñ+ùë†ùëó —è–≤–ª—è–µ—Ç—Å—è –ø—Ä–∞–≤–∏–ª—å–Ω–æ–π —Å–∫–æ–±–æ—á–Ω–æ–π –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å—é.

–ü—Ä–∏–º–µ—Ä—ã\
–≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\
3\
)\
()\
(\
–≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\
2

```python
def bracket_check(text):
    stack = []
    for ch in text:
        if ch == '(':
            stack.append('(')
        elif ch == ')':
            if not stack or stack[-1] != '(':
                return False
            stack.pop()
    if not stack:
        return True
    else:
        return False

n = int(input())
if n < 1 or n > 315:
    print("n is out of range")
    exit()
s = []
count = 0
for i in range(n):
    s.append(input())
for i in range(n):
    for j in range(n):
        if bracket_check(s[i] + s[j]):
            count += 1
print(count)
```

2

```python
n, m = int(input()), list(map(int, input().split()))
if n < 1 or n > 105:
    print("n is out of range")
    exit()
for item in m:
    if item < 0 or item > 208:
        print("m is out of range")
        exit()
a, b = sum(m) // n, (sum(m) + n - 1) // n
print(max(sum(a - x for x in m if x < a), sum(x - b for x in m if x > b)))
```

```python
n = int(input())
if n < 1 or n > 1000:
    print("n is out of range")
    exit()
s = [int(c) for c in input()]
ans = [10]
for i in range(n):
    s = s[1:] + [s[0]]
    pwd = [0]
    for j in range(1, n):
        d = (pwd[-1] + s[j] - s[j-1]) % 10
        pwd.append(d)
    if ans > pwd:
        ans = pwd
for d in ans:
    print(d, end = '')
```
