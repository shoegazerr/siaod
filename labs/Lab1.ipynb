### Задание №3
Реализовать методы сортировки строк числовой матрицы в соответствии с

временем стандартной функции сортировки. Испытания проводить на сгенерированных

```python
import copy
```

```python
# Сортировка выбором.
# Берётся срез массива, в котором минимальный элемент переносят в самый левый угол,
# после чего срез уменьшается и цикл повторяется.
import time
start_time = time.time()
def select(arr):
    for i in range (len(arr)):
        a=arr[i]
        for j in range (i+1,len(arr)):
            if arr[j]<arr[i]:
                arr[i]=arr[j]
                arr[j]=a
for arr in matrix:
    arr=select(arr)
for i in range(user_m):
    print(matrix[i])
select_time = round((time.time() - start_time)*1000)
print("--- {0} ms ---".format(select_time))
```

```python
# Сортировка вставкой.
start_time = time.time()
def paste(arr):
    for i in range(len(arr)):
        a = arr[i]
        position = i
        
        while position > 0 and arr[position - 1] > a:
            arr[position] = arr[position - 1]
            position = position - 1
        arr[position] = a
    return arr
for arr in matrix:
    arr=paste(arr)
for i in range(user_m):
    print(matrix[i])
paste_time = round((time.time() - start_time)*1000)
print("--- {0} ms ---".format(paste_time))
```

```python
# Сортировка обменом. || Сортировка пузырьком.
start_time = time.time()
def trade(arr):
    for i in range(0,len(arr)-1):
        a = arr[i]
        if arr[i]>arr[i+1]:
            arr[i] = arr[i+1]
            arr[i+1] = a
    return arr
for arr in matrix:
    arr=trade(arr)
for i in range(user_m):
    print(matrix[i])
trade_time = round((time.time() - start_time)*1000)
print("--- {0} ms ---".format(trade_time))
```

```python
# Сортировка Шелла.
start_time = time.time()
def shell(arr):
    hf=len(arr) // 2
    while hf > 0:
        for i in range(hf, len(arr)):
            t = arr[i]
            j = i
            while j >= hf and arr[j - hf] > t:
                arr[j] = arr[j -  arr]
                j -=  hf
            arr[j] = t
        hf = hf // 2
for arr in matrix:
    arr=shell(arr)
for i in range(user_m):
    print(matrix[i])
shell_time = round((time.time() - start_time)*1000)
print("--- {0} ms ---".format(shell_time))
```

```python
# Быстрая сортировка.
start_time = time.time()
def partition(arr, start, end):
    pivot = arr[start]
    left = start + 1
    right = end
    while True:
        while arr[right] > pivot and right > start:
            right -= 1
        while arr[left] <= pivot and left < end:
            left += 1
        if left >= right:
            break
        temp = arr[left]
        arr[left] = arr[right]
        arr[right] = temp
    arr[start] = arr[right]
    arr[right] = pivot
    return right
def __quick_sort(arr, start, end):
    if end > start:
        pivot_index = partition(arr, start, end)
        __quick_sort(arr, start, pivot_index - 1)
        __quick_sort(arr, pivot_index + 1, end)
def quick_sort(arr):
    __quick_sort(arr, 0, len(arr) - 1)  
for arr in matrix:
    arr=quick_sort(arr)
for i in range(user_m):
    print(matrix[i])
quick_sort_time = round((time.time() - start_time)*1000)
print("--- {0} ms ---".format(quick_sort_time))
```

```python
# Турнирная сортировка.
start_time = time.time()
import heapq
def sort_tournament(line, places):
    heap = []
    if len(line) >= places:
        heap = line[:places]
    else:
        heap = line[:]
    heapq.heapify(heap)
    winners = [float('-inf')]
    losers = []
    i = places
    while i < len(line) or len(heap) > 0:
        heap_top = heapq.heappop(heap)
        if heap_top >= winners[-1]:
            winners.append(heap_top)
        else:
            losers.append(heap_top)
        if i < len(line):                       
            heapq.heappush(heap, line[i])       
        i += 1
    del winners[0]
    line.clear()
    line.extend(losers)
    line.extend(winners)
    if len(losers) > 0:
        sort_tournament(line, places)
for i in range(len(matrix)):
    sort_tournament(matrix[i], 4)
for i in range(user_m):
    print(matrix[i])
sort_tournament_time = round((time.time() - start_time)*1000)
print("--- {0} ms ---".format(sort_tournament_time))
```
