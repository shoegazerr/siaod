```python
#генерация
import copy
import time
from random import randint

length=int(input('Длина: '))
min_value=int(input('Мин: '))
max_value=int(input('Макс: '))
randata=[]
for i in range(length):
    randata.append(randint(min_value,max_value))
print(randata)
#Бинарный поиск
def bin_search(randata, val):
    randata.sort()
    first = 0
    last = len(randata)-1
    index = -1
    while (first <= last) and (index == -1):
        mid = (first+last)//2
        if randata[mid] == val:
            index = mid
        else:
            if val<randata[mid]:
                last = mid -1
            else:
                first = mid +1
    return index
def bin_delete(randata, val):
    val_pos = bin_search(randata,val) 
    if val_pos != -1:
        randata.pop(val_pos)
        return True
    else:
        return False    
def bin_append(randata, val):
    if len(randata) == 1:
        if randata[0] < val:
            randata.append(val)
        else:
            randata.insert(0, val)
        return
    first=0
    last=len(randata)-1
    while (first<=last): 
        mid = (first+last)//2
        if randata[mid]<=val and randata[mid+1]>=val:
            randata.insert(mid+1,val)
            return randata
        else:
            if val<randata[mid]:
                last=mid-1
            else:
                first=mid+1
randata1=copy.deepcopy(randata)
start_time=time.time()
print(bin_search(randata1,15))
bin_search_time=round((time.time()-start_time)*1000)
randata1.sort()
print(randata1)
print(bin_delete(randata1,10))
print(bin_append(randata1,20))
print("--- {0} ms ---".format(bin_search_time))
```

```python
#Бинарное дерево
class Node:
    def __init__(self, key):
        self.key = key
        self.left = None
        self.right = None
# Отсортированный (inorder) обход
def inorder(root):
    if root is not None:
        # Обходим левое поддерево
        inorder(root.left)
        # Обходим корень
        print(str(root.key) + "->", end=' ')
        # Обходим правое поддерево
        inorder(root.right)
# Вставка элемента
def insert(node, key):
    # Возвращаем новый узел, если дерево пустое
    if node is None:
        return Node(key)

    # Идем в нужное место и вставляет узел
    if key < node.key:
        node.left = insert(node.left, key)
    else:
        node.right = insert(node.right, key)
    return node
# Поиск inorder-преемника
def minValueNode(node):
    current = node
    # Найдем крайний левый лист — он и будет inorder-преемником
    while(current.left is not None):
        current = current.left
    return current
# Удаление узла
def deleteNode(root, key):
    # Возвращаем, если дерево пустое
    if root is None:
        return root
    # Найдем узел, который нужно удалить
    if key < root.key:
        root.left = deleteNode(root.left, key)
    elif(key > root.key):
        root.right = deleteNode(root.right, key)
    else:
        # Если у узла только один дочерний узел или вообще их нет        
        if root.left is None:
            temp = root.right
            root = None
            return temp
        elif root.right is None:
            temp = root.left
            root = None
            return temp
        # Если у узла два дочерних узла,
        # помещаем центрированного преемника
        # на место узла, который нужно удалить
        temp = minValueNode(root.right)
        root.key = temp.key
        # Удаляем inorder-преемниа
        root.right = deleteNode(root.right, temp.key)
    return root
root = None
root = insert(root, 8)
root = insert(root, 3)
root = insert(root, 1)
root = insert(root, 6)
root = insert(root, 7)
root = insert(root, 10)
root = insert(root, 14)
root = insert(root, 4)
print("Начальное дерево: ", end=' ')
inorder(root)
print("\nПосле удаления 10")
root = deleteNode(root, 10)
print("Проверка: ", end=' ')
inorder(root)
print("\nПосле добавления 13")
root = insert(root, 13)
print("Проверка: ", end=' ')
inorder(root)
```
